```python
import wikipediaapi
import pandas as pd
```

```python
df_wiki = pd.read_csv("../data/wiki_results.csv")
```

```python
df_wiki.groupby('genderLabel').value_counts(normalize=True)
```

```python
df_wiki["title"] = df_wiki["article"].str.split("/").str[-1]
df_wiki["title"] = df_wiki["title"].str.replace("_", " ")
```

```python
wiki = wikipediaapi.Wikipedia(
    language='de',
    user_agent='research-project'
)

data = []

for title in df_wiki["title"]:
    page = wiki.page(title)

    if page.exists():
        data.append({
            "title": title,
            "text": page.text
        })

df = pd.DataFrame(data)

print(df["text"][0][:1000])
```

```python
df_wiki = df_wiki.rename(columns={
    "personLabel": "name"
})

df = df.rename(columns={
    "title": "name"
})
```

```python
df = df_wiki.merge(df, on="name", how="inner")
```
